```python
class Bpe_algorithm:
    def __init__(self,total_vocab_len):
        self.bpe_vocab=[]
        self.nospace_token=[]
        self.vocab_len=total_vocab_len
        
    def read_wordlist(self):
        unique_word=[]
        with open('wordlist.txt', 'r', encoding='utf-8') as file:
            for line in file:
                charac=[]
                for character in line:
                    if(character!='\n'):
                        unique_word.append(character)
                        charac.append(character)
                self.nospace_token.append(charac)
        self.bpe_vocab=list(set(unique_word))
        
    def merge_word(self):
        dic={}
        for ch in self.nospace_token:
            for i in range(len(ch)-1):
                key=ch[i]+ch[i+1]
                if key in dic:
                    dic[key] += 1
                else:
                    dic[key] = 1

        sorted_dict_descending = dict(sorted(dic.items(), key=lambda item: item[1], reverse=True))
#         print(sorted_dict_descending)

        if sorted_dict_descending:
            first_key = next(iter(sorted_dict_descending))
        self.bpe_vocab.append(first_key)
        return first_key
    
    
    def merge_word_corpus(self,key):
        for ch in self.nospace_token:   #### corpus character pick
            for i in range(len(ch) - 1):
                if i + 1 < len(ch) and (ch[i] + ch[i+1] == key):
                    ch[i] = key
                    ch.pop(i+1)
                    i -= 1
                    
                    
    def test_bpe(self):
        uni_words=[]
        with open('50_nospaces.txt', 'r', encoding='utf-8') as file:
            for line in file:
                charac=[]
                for character in line:
                    if(character!='\n'):
                        charac.append(character)
                        
                uni_words.append(charac)

        ##### reverse bpe_vocab to find max length string
        sorted_list = sorted(self.bpe_vocab, key=len, reverse=True)
        length=len(sorted_list[0])
        
        ##### run until max length character join
        for k in range(length):
            for uni in uni_words:      
                for i in range(len(uni)-1):
                    if(i+1<len(uni)):
                        word=uni[i]+uni[i+1]
                        bool=False
                        for j in self.bpe_vocab: 
                            if(word==j ):
                                bool=True
                        if(bool==True):
                            uni[i]=word
                            uni.pop(i+1)
                            i -= 1
                        
#         print(uni_words)
        
        with open("output.txt", "w", encoding="utf-8") as file:
            for word in uni_words:
                print(word)
                st=""
                for w in word:
                    st=st+w+" "
                file.write(st+"\n")

                 
    def BPE_Algorithm(self):
        self.read_wordlist()
        while len(self.bpe_vocab) <= self.vocab_len:
            key=self.merge_word()
            self.merge_word_corpus(key)
            
#         print("bpe_vocab",self.bpe_vocab)
#         print("nospace",self.nospace_token)
        
        
if __name__ == "__main__":
    obj=Bpe_algorithm(1000)
    obj.BPE_Algorithm()
    obj.test_bpe()
                 
```
